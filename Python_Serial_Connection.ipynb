```python
import serial
import tkinter as tk
import threading

# Serial configuration and its function
port = 'COM6'  # Replace/change according to the COM port used.
baud_rate = 9600

def read_serial():
    while window_exists:
        try:
            response = ser.readline().decode().strip()
            if response:
                text.insert(tk.END, response + '\n')
                text.see(tk.END)  
        except serial.serialutil.SerialException:
            break

# Function to send data to the serial connection
def send_data(event=None):
    global input_data
    data = entry.get()
    if data:
        ser.write(data.encode())
        input_data = data
        entry.delete(0, tk.END)

# Create the main window and scroolbar
window = tk.Tk()
window.title("Serial Reader")

text = tk.Text(window, height=20, width=120)
text.pack()

scrollbar = tk.Scrollbar(window)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

text.config(yscrollcommand=scrollbar.set)
scrollbar.config(command=text.yview)

input_data = "" # Fill it with certain data if needed
entry = tk.Entry(window, width=50)
entry.pack()
entry.insert(tk.END, input_data)

# Open the serial connection and reading response
ser = serial.Serial(port, baud_rate)

window_exists = True
serial_thread = threading.Thread(target=read_serial)
serial_thread.daemon = True  
serial_thread.start()

entry.bind("<Return>", send_data)

# Function to setting the Window
def close_window():
    global window_exists
    window_exists = False
    ser.close()
    window.destroy()

window.protocol("WM_DELETE_WINDOW", close_window)

window.mainloop()
```
